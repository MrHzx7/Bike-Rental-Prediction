```python
import pandas as pd
import matplotlib.pyplot as plt

# Define file paths
nestle_file = r"C:\Users\haris\Desktop\historical data\Nestle_India_Historical_Data_2015_2023.xlsx"
britannia_file = r"C:\Users\haris\Desktop\historical data\Britannia_Historical_Data_2015_2023.xlsx"

# Load data into DataFrames, specifying the correct header row
nestle_data = pd.read_excel(nestle_file, header=2)  # Third row as header
britannia_data = pd.read_excel(britannia_file, header=2)

# Print column names to verify structure
print("Nestlé Data Columns:", nestle_data.columns)
print("Britannia Data Columns:", britannia_data.columns)

# Rename columns based on the actual number of columns in your dataset
nestle_data.columns = ['Date', 'Price', 'Close', 'High', 'Low', 'Open']
britannia_data.columns = ['Date', 'Price', 'Close', 'High', 'Low', 'Open']

# Ensure the 'Date' column is in datetime format and set as index
nestle_data['Date'] = pd.to_datetime(nestle_data['Date'])
britannia_data['Date'] = pd.to_datetime(britannia_data['Date'])
nestle_data.set_index('Date', inplace=True)
britannia_data.set_index('Date', inplace=True)

# Plot the Closing Prices of Both Stocks Over Time
plt.figure(figsize=(14, 7))

# Plot Nestlé data
plt.plot(nestle_data['Close'], label="Nestlé India", color='blue')

# Plot Britannia data
plt.plot(britannia_data['Close'], label="Britannia Industries", color='green', linestyle='dashed')

# Add title, labels, legend, and grid
plt.title('Stock Prices of Nestlé India and Britannia (2015-2023)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (INR)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate and Plot Moving Averages for Each Stock (50-day and 200-day)
# Create a copy of the data to calculate moving averages
moving_averages_nestle = nestle_data[['Close']].copy()
moving_averages_britannia = britannia_data[['Close']].copy()

# Calculate moving averages for Nestlé
moving_averages_nestle['50_MA'] = moving_averages_nestle['Close'].rolling(window=50).mean()
moving_averages_nestle['200_MA'] = moving_averages_nestle['Close'].rolling(window=200).mean()

# Calculate moving averages for Britannia
moving_averages_britannia['50_MA'] = moving_averages_britannia['Close'].rolling(window=50).mean()
moving_averages_britannia['200_MA'] = moving_averages_britannia['Close'].rolling(window=200).mean()

# Plot the moving averages
plt.figure(figsize=(14, 7))

# Nestlé moving averages plot
plt.plot(moving_averages_nestle.index, moving_averages_nestle['Close'], label="Nestlé Close", color='blue', alpha=0.5)
plt.plot(moving_averages_nestle.index, moving_averages_nestle['50_MA'], label="Nestlé 50-day MA", color='blue', linestyle='dashed')
plt.plot(moving_averages_nestle.index, moving_averages_nestle['200_MA'], label="Nestlé 200-day MA", color='blue', linestyle='dotted')

# Britannia moving averages plot
plt.plot(moving_averages_britannia.index, moving_averages_britannia['Close'], label="Britannia Close", color='green', alpha=0.5)
plt.plot(moving_averages_britannia.index, moving_averages_britannia['50_MA'], label="Britannia 50-day MA", color='green', linestyle='dashed')
plt.plot(moving_averages_britannia.index, moving_averages_britannia['200_MA'], label="Britannia 200-day MA", color='green', linestyle='dotted')

# Add title, labels, legend, and grid
plt.title('50-day and 200-day Moving Averages of Nestlé and Britannia (2015-2023)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (INR)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define file paths
nestle_file = r"C:\Users\haris\Desktop\historical data\Nestle_India_Historical_Data_2015_2023.xlsx"
britannia_file = r"C:\Users\haris\Desktop\historical data\Britannia_Historical_Data_2015_2023.xlsx"

# Load data into DataFrames, specifying the correct header row
nestle_data = pd.read_excel(nestle_file, header=2)  # Third row as header
britannia_data = pd.read_excel(britannia_file, header=2)

# Print column names to verify structure
print("Nestlé Data Columns:", nestle_data.columns)
print("Britannia Data Columns:", britannia_data.columns)

# Rename columns based on the actual number of columns in your dataset
nestle_data.columns = ['Date', 'Price', 'Close', 'High', 'Low', 'Open']
britannia_data.columns = ['Date', 'Price', 'Close', 'High', 'Low', 'Open']

# Ensure the 'Date' column is in datetime format and set as index
nestle_data['Date'] = pd.to_datetime(nestle_data['Date'])
britannia_data['Date'] = pd.to_datetime(britannia_data['Date'])
nestle_data.set_index('Date', inplace=True)
britannia_data.set_index('Date', inplace=True)

# Plot the Closing Prices of Both Stocks Over Time
plt.figure(figsize=(14, 7))

# Plot Nestlé data
plt.plot(nestle_data['Close'], label="Nestlé India", color='blue')

# Plot Britannia data
plt.plot(britannia_data['Close'], label="Britannia Industries", color='green', linestyle='dashed')

# Add title, labels, legend, and grid
plt.title('Stock Prices of Nestlé India and Britannia (2015-2023)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (INR)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()

# Identify the Highest and Lowest Closing Prices for each stock
stocks = {'Nestlé India': nestle_data, 'Britannia Industries': britannia_data}

for stock_name, stock_data in stocks.items():
    highest_price = stock_data['Close'].max()
    lowest_price = stock_data['Close'].min()
    highest_date = stock_data['Close'].idxmax()
    lowest_date = stock_data['Close'].idxmin()
    
    print(f"\n{stock_name}:")
    print(f"Highest Closing Price: ₹{highest_price:.2f} on {highest_date.date()}")
    print(f"Lowest Closing Price: ₹{lowest_price:.2f} on {lowest_date.date()}")
```

```python
# Calculate Daily Returns for both stocks
nestle_daily_returns = nestle_data['Close'].pct_change()
britannia_daily_returns = britannia_data['Close'].pct_change()

# Combine daily returns into a single DataFrame for plotting
daily_returns = pd.DataFrame({
    'Nestlé India': nestle_daily_returns,
    'Britannia Industries': britannia_daily_returns
})

# Plot the distribution of daily returns for both stocks
plt.figure(figsize=(10, 6))
for stock_name in daily_returns.columns:
    sns.histplot(daily_returns[stock_name].dropna(), bins=100, kde=True, label=stock_name)

# Add title, labels, legend, and grid
plt.title('Distribution of Daily Returns for Nestlé India and Britannia', fontsize=16)
plt.xlabel('Daily Return', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate Cumulative Returns for both stocks
cumulative_returns = (1 + daily_returns).cumprod()

# Plot the cumulative returns for both stocks
plt.figure(figsize=(14, 7))
for stock_name in cumulative_returns.columns:
    plt.plot(cumulative_returns[stock_name], label=stock_name)

# Add title, labels, legend, and grid
plt.title('Cumulative Returns of Nestlé India and Britannia (2015-2023)', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Cumulative Return', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate 30-day Rolling Standard Deviation (Volatility) for both stocks
rolling_std = daily_returns.rolling(window=30).std()

# Plot the rolling standard deviation for both stocks
plt.figure(figsize=(14, 7))
for stock_name in rolling_std.columns:
    plt.plot(rolling_std[stock_name], label=stock_name)

# Add title, labels, legend, and grid
plt.title('Rolling 30-Day Volatility of Nestlé India and Britannia', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Volatility (Standard Deviation)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Calculate the Correlation Matrix between Nestlé India and Britannia
correlation_matrix = daily_returns.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Add title for the heatmap
plt.title('Correlation Matrix between Nestlé India and Britannia', fontsize=16)

# Show the heatmap
plt.show()
```
